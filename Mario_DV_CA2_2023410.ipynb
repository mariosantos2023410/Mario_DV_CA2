#### The differences between a Graph and a Tree
A Graph and a Tree have some similar and different points. They are both non-linear data structure and also both of them have a collection of nodes and edges. However, a Tree has a single node called root and can have various child nodes but is unable to form a cycle. On the other hand, a Graph doesn't have a single node(root), which means any node in the Graph can be considered as a startint point. Additionaly, a Graph can have various edges and form cycles.
