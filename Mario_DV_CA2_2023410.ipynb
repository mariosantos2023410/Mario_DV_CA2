#### The differences between a Graph and a Tree
A Graph and a Tree have some similar and different points. They are both non-linear data structure and also both of them have a collection of nodes and edges. However, a Tree has a single node called root and can have various child nodes but is unable to form a cycle. On the other hand, a Graph doesn't have a single node(root), which means any node in the Graph can be considered as a startint point. Additionaly, a Graph can have various edges and form cycles.

#### Shortest Path as an abstraction of the problem

Given the figure 1, the abstraction of the problem here is to find the shortest path between the directions starting point A travelling to the end solution 
which is S.To find the shortest path to between this two nodes, this can be calculated using an algorithm like Djikstra's or A* where they calculate

On the other hand A* algorithm uses heuristics function which guides to the short path and also estimates the distance from one node to another.

#### Implementation of Dijkstra's Algorithm

```python
import heapq

def calc_dist(puzzle_graph, start_node):

    dist = {node: float('infinity') for node in puzzle_graph}

    dist[start_node] = 0

    pq = [(0, start_node)]

    while pq:
        ongoing_dist, ongoing_node = heapq.heappop(pq)

        if ongoing_dist > dist[ongoing_node]:
            continue

        for next_node, weight in puzzle_graph[ongoing_node].items():
            new_dist = ongoing_dist + weight

            if new_dist < dist[next_node]:
                dist[next_node] = new_dist
                heapq.heappush(pq, (new_dist, next_node))

    return dist
```

#### Implementation of A* Algorithm

```python
from collections import deque

class Puzzle:
    def __init__(self, puzzle_graph):
        self.puzzle_graph = puzzle_graph

    def get_nodes(self, node):
        return self.puzzle_graph[node]

    def estimate_cost(self, node):
        H = 1

        return H

    def find_shortest_path(self):
        start = 'A'
        goal = 'S'

        open_set = set([start])
        closed_set = set([])

        costs = {}
        costs[start] = 0

        parents = {}
        parents[start] = start

        while open_set:
            current_node = None

            
            for node in open_set:
                if current_node is None or costs[node] + self.estimate_cost(node) < costs[current_node] + self.estimate_cost(current_node):
                    current_node = node

            if current_node is None:
                print('No path found!')
                return None

            
            if current_node == goal:
                path = []

                while parents[current_node] != current_node:
                    path.append(current_node)
                    current_node = parents[current_node]

                path.append(start)
                path.reverse()

                print('Path found:', path)
                return path

          
            for (neighbor, weight) in self.get_nodes(current_node):
                if neighbor not in open_set and neighbor not in closed_set:
                    open_set.add(neighbor)
                    parents[neighbor] = current_node
                    costs[neighbor] = costs[current_node] + weight
                else:
                    if costs[neighbor] > costs[current_node] + weight:
                        costs[neighbor] = costs[current_node] + weight
                        parents[neighbor] = current_node

                        if neighbor in closed_set:
                            closed_set.remove(neighbor)
                            open_set.add(neighbor)

            
            open_set.remove(current_node)
            closed_set.add(current_node)

        print('No path found!')
        return None
```
